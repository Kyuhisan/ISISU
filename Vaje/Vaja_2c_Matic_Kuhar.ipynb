```python
# 1. Naložite nasledenje podatke v nov notebook (.ipynb) ali .py dokument:
import pandas as pd

DataFrame1 = pd.read_csv("/content/drive/MyDrive/university_rank_2020.csv", sep=",", index_col="University", decimal=".")
DataFrame2 = pd.read_csv("/content/drive/MyDrive/university_rank_other.csv", sep=" ", index_col=0, decimal=".")
DataFrame3 = pd.read_excel("/content/drive/MyDrive/university_rank_students.xlsx", index_col="University")

# DataFrame1.head(5)
DataFrame2.head(5)
# DataFrame3.head(5)
```

```python
# 2. Vse DataFrame-e združite v en DataFrame (poimenujte ga df).
df = pd.concat([DataFrame1, DataFrame2, DataFrame3], axis=1)
# df.head(25)

df.shape
```

```python
# 3. Izpišite število vrstic in stolpcev, ki jih ima novo ustvarjeni df in preglejte kako izgleda prvih 10 vrstic.
df.shape
df.head(10)
```

```python
# 4. Preverite koliko je manjkajočih podatkov v posameznem stolpcu.
df.isnull().sum()
```

```python
# 5. Manjkajoče vrednosti za stolpec International_Outlook ročno dopolnite glede na stolpec Teaching
# - če je vrednost v posamezni vrstici za atribut Teaching:
# < 20 - dopolnite stolpec International_Outlook z vrednostjo "bad"
# > 60  - dopolnite stolpec International_Outlook z vrednostjo "excellent"
# vmesne vrednosti pa dopolnite z vrednostjo "good"

df.loc[df.Teaching < 20, "International_Outlook"] = "bad"
df.loc[df.Teaching > 60, "International_Outlook"] = "excellent"
df.loc[(df.Teaching >= 20) & (df.Teaching <= 60), "International_Outlook"] = "good"
df.International_Outlook
```

```python
# 6. Manjkajoče vrednosti v ostalih stolpcih dopolnite s povprečjem stolpca.
df.fillna(df["International_Students(%)"].mean(), inplace=True)
df.fillna(df["Percentage_Female(%)"].mean(), inplace=True)
df.fillna(df["Percentage_Male(%)"].mean(), inplace=True)

df.isnull().sum()
```

```python
# 7. Izpišite katere unikatne vrednosti obstajajo za atribut International_Outlook in koliko je primerkov posamezne kategorije.
df.International_Outlook.value_counts()
```

```python
# 8. Izpišite koliko je univerz v posamezni državi.
df.Country.value_counts()
```

```python
# 9. Izpišite vse univerze, ki se nahajajo v Sloveniji.
df.loc[df.Country == "Slovenia"]
```

```python
# 10. Ustvarite nov stolpec poimenovan Staff Skills, ki se naj izračuna kot vsota stolpcev Teaching in Research,
# deljeno z 2  (Teaching + Research/ 2). Izpišite prve tri vrstice in preverite izračunane vrednosti.

df["Staff Skills"] = (df.Teaching + df.Research) / 2
df.head(3)
```

```python
# 11. Izbrišite stolpec Staff Skills.
df.drop("Staff Skills", axis=1, inplace=True)
```

```python
# 12. DataFrame df shranite v csv kot "university_rank.csv" s seperatorjem ; in decimalnim znakom pika.
df.to_csv("university_rank.csv", sep=";", decimal=".")
```
