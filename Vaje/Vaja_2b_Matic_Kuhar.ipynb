```python
# Manjkajoče številske podatke nadomestite s povprečjem.
dataframe["Car Year"].fillna(dataframe["Car Year"].mean(), inplace=True)
dataframe["Movie Year"].fillna(dataframe["Movie Year"].mean(), inplace=True)
dataframe.isnull().sum()
```

```python
# Manjkajoče kategorične podatke ustrezno nadomestite.
dataframe["Movie Years and Episodes"].fillna("Still Running", inplace=True)
dataframe["IMDb Link"].fillna("Link Not Found", inplace=True)
dataframe["Class"].fillna("Vehicle", inplace=True)
dataframe["Car Type"].fillna("Vehicle", inplace=True)
dataframe["Car Origin"].fillna("Unknown", inplace=True)
dataframe["Car Built In"].fillna("Unknown", inplace=True)
dataframe["Car Made For"].fillna("Unknown", inplace=True)
dataframe["Time In Movie"].fillna("Unknown", inplace=True)
dataframe["Car Image Link"].fillna("Image Not Found", inplace=True)
dataframe["IMDb Movie Id"].fillna("Not Found", inplace=True)
dataframe.isnull().sum()
```

```python
# Smiselno uporabite grupiranje in slekcijo stolpcev.
# Izpis znamke in leta v katerem se je ta znamka pojavila največkrat
dataframe.groupby("Brand")["Movie Year"].max()
```
